# Data Description
- This dataset includes hourly and daily bike rental counts for Capital Bikeshare from 2011 and 2012, with weather and seasonal details.
- There are 8645 rows.
- There are 15 variables:
    - season: Season (Winter=1, Spring=2, Summer=3, Fall=4)
    - mnth: Month
    - day: Day of the year (1 to 365)
    - hr: Hour of the day (0 to 23)
    - holiday: Holiday (Yes=1, No=0)
    - weekday: Day of the week (Sunday=0 to Saturday=6)
    - workingday: Workday (Yes=1, No=0)
    - weathersit: Weather condition
    - temp: Normalized temperature (Celsius)
    - atemp: Normalized "feels like" temperature (Celsius)
    - hum: Normalized humidity
    - windspeed: Normalized wind speed
    - casual: Casual bikers count
    - registered: Registered bikers count
    - bikers: Total bikers count

# Load Packages and Data

```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
import statsmodels.api as sm
from ISLP import load_data
from ISLP.models import (ModelSpec as MS, summarize)
from ISLP import confusion_table
from ISLP.models import contrast
from sklearn.discriminant_analysis import \
     (LinearDiscriminantAnalysis as LDA,
      QuadraticDiscriminantAnalysis as QDA)
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
Bike = load_data('Bikeshare')
Bike.head()
```

```python
Bike.describe().round(1)
```

```python
Bike.weathersit.value_counts()
```

# Linear Regression

```python
X = MS(['mnth','hr','workingday','temp','weathersit']).fit_transform(Bike)
Y = Bike['bikers']
M_lm = sm.OLS(Y, X).fit()
summarize(M_lm)
```

- `hr` has 24 levels and there are 40 predictors in total.
- `hr[0]` and `mnth[Jan]` are baseline values with zero coefficients; other levels are measured against these.
- For example, the February coefficient of 6.845 means about 7 more riders in February than January, and March has about 16.5 more riders than January.

```python
hr_encode = contrast('hr', 'sum')
mnth_encode = contrast('mnth', 'sum')
```

- Refitting again:

```python
X2 = MS([mnth_encode,hr_encode,'workingday','temp','weathersit']).fit_transform(Bike)
M2_lm = sm.OLS(Y, X2).fit()
S2 = summarize(M2_lm); S2
```

- All but `hr[23]` and `mnth[Dec]` have reported coefficients.
- The unreported coefficients for the last levels are the negative sum of all other levels, making the coefficients sum to zero and represent differences from the mean.
- For example, January's coefficient of $-46.087$ means 46 fewer riders in January compared to the yearly average.
- The choice of coding doesn't matter as long as the model output is interpreted correctly, yielding the same predictions regardless of coding.

```python
np.sum((M_lm.fittedvalues - M2_lm.fittedvalues)**2)
```

- We plot the coefficient estimates for `mnth` on the left and for `hr` on the right.
- The January to November coefficients are available. The December coefficient is the negative sum of the others.

```python
coef_month = S2[S2.index.str.contains('mnth')]['coef']
coef_month
```

- We append `Dec` as the negative of the sum of all other months.

```python
months = Bike['mnth'].dtype.categories
coef_month = pd.concat([coef_month, pd.Series([-coef_month.sum()], index=['mnth[Dec]'])])
coef_month
```

- We redo the similar process for `hr`

```python
coef_hr = S2[S2.index.str.contains('hr')]['coef']
coef_hr = coef_hr.reindex(['hr[{0}]'.format(h) for h in range(23)])
coef_hr = pd.concat([coef_hr,
                     pd.Series([-coef_hr.sum()], index=['hr[23]'])
                    ])
```

```python
# Set the plot up
fig, (ax_month, ax_hr) = plt.subplots(1, 2, figsize=(16, 8))
# First subplot for month
x_month = np.arange(coef_month.shape[0])
ax_month.plot(x_month, coef_month, marker='o', ms=10)
ax_month.set_xticks(x_month)
ax_month.set_xticklabels([l[5] for l in coef_month.index])
ax_month.set_xlabel('Month'); ax_month.set_ylabel('Coefficient')
# Second subplot for hour
x_hr = np.arange(coef_hr.shape[0])
ax_hr.plot(x_hr, coef_hr, marker='o', ms=10)
ax_hr.set_xticks(x_hr[::2])
ax_hr.set_xticklabels(range(24)[::2])
ax_hr.set_xlabel('Hour'); ax_hr.set_ylabel('Coefficient');
```

# Poisson Regression

```python
M_pois = sm.GLM(Y, X2, family=sm.families.Poisson()).fit()
```

- We redo the similar process to plot for `mnth` and `hr`

```python
S_pois = summarize(M_pois)
coef_month = S_pois[S_pois.index.str.contains('mnth')]['coef']
coef_month = pd.concat([coef_month,
                        pd.Series([-coef_month.sum()],
                                   index=['mnth[Dec]'])])
coef_hr = S_pois[S_pois.index.str.contains('hr')]['coef']
coef_hr = pd.concat([coef_hr,
                     pd.Series([-coef_hr.sum()],
                     index=['hr[23]'])])
```

```python
# Set the plot up
fig, (ax_month, ax_hr) = plt.subplots(1, 2, figsize=(16, 8))
# First subplot for month
x_month = np.arange(coef_month.shape[0])
ax_month.plot(x_month, coef_month, marker='o', ms=10)
ax_month.set_xticks(x_month)
ax_month.set_xticklabels([l[5] for l in coef_month.index])
ax_month.set_xlabel('Month'); ax_month.set_ylabel('Coefficient')
# Second subplot for hour
x_hr = np.arange(coef_hr.shape[0])
ax_hr.plot(x_hr, coef_hr, marker='o', ms=10)
ax_hr.set_xticks(x_hr[::2])
ax_hr.set_xticklabels(range(24)[::2])
ax_hr.set_xlabel('Hour'); ax_hr.set_ylabel('Coefficient');
```

- We plot fitted values of two models.

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax.scatter(M2_lm.fittedvalues,
           M_pois.fittedvalues,
           s=2)
ax.set_xlabel('Linear Regression Fit')
ax.set_ylabel('Poisson Regression Fit')
ax.axline([0,0], c='black', linewidth=3, linestyle='--', slope=1);
```

- Poisson regression predictions are correlated with linear model predictions but are non-negative and tend to be larger for very low or high ridership levels.
- We fit Poisson regression models using `sm.GLM(family=sm.families.Poisson())`. Previously, we used `sm.GLM(family=sm.families.Binomial())` for logistic regression. Other `family` options, like `sm.families.Gamma()`, fit different GLMs.
